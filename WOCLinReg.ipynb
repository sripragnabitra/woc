```python
#calculating the function
def f_wb(w,b):
  return w * x_train + b 
```

```python
#calculating cost function
m = x_train.shape[0] 
def j_wb(x_train, y_train, w, b):
  return (1/ (2 * m)) * np.sum(np.square(f_wb(w,b)-y_train))
```

```python
#calculating cost function
def j_wb(x_train, y_train, w, b):  
    m = x_train.shape[0] 
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x_train[i] + b  
        cost = (f_wb - y_train[i]) ** 2  
        cost_sum = cost_sum + cost  #or cost_sum += cost
    j_wb = (1 / (2 * m)) * cost_sum  

    return j_wb
```

```python
#calculating gradients
def gradient_function(x_train, y_train, w, b): 
    m = x_train.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):  
        f_wb = w * x_train[i] + b 
        dj_dw_i = (f_wb - y_train[i]) * x_train[i] 
        dj_db_i = f_wb - y_train[i] 
        dj_db += dj_db_i
        dj_dw += dj_dw_i 
    dj_dw = dj_dw / m 
    dj_db = dj_db / m 
        
    return dj_dw, dj_db
```

```python
#calculating gradient descent
import math
def gradient_descent(x_train, y_train, w_in, b_in, alpha, num_iters, j_wb, gradient_function): #alpha is learning rate
    J_history = []
    p_history = [] #arrays to store the coming values
    b = b_in
    w = w_in
    
    for i in range(num_iters):
        dj_dw, dj_db = gradient_function(x_train, y_train, w , b)     
        b = b - alpha * dj_db                            
        w = w - alpha * dj_dw                            
        if i<100:      #not only 100; can be any number
            J_history.append(j_wb(x_train, y_train, w , b))
            p_history.append([w,b]) #saving cost and parameters into the arrays
        if i% math.ceil(num_iters/10) == 0:   # Print cost and others at intervals of 10% of num_iters like if num_iters is 500 intervals of 50 are printed so 0,50,100,150 and so on
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
 
    return w, b, J_history, p_history
```

```python
#finding suitable w and b
w_in = 0
b_in = 0
num_iters = 500
tmp_alpha = 1.0e-6
w_final, b_final, J_hist, p_hist = gradient_descent(x_train ,y_train, w_in, b_in, tmp_alpha, num_iters, j_wb, gradient_function)
print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
print('minimum cost is: ', j_wb(x_train, y_train, w_final, b_final))
```
