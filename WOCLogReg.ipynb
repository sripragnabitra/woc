# **LOGISTIC REGRESSION**

**TRAINING MY MODEL**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/sripragnabitra/woc/main/classification_train.csv')
df
```

```python
#for x_train, I took 0 to 199 index rows of pixel 60 to pixel 79 columns for logistic regression
x_train=df.iloc[:200,61:81]  
x_train=np.array(x_train)
print(x_train)
print('number of training features:', len(x_train))
```

```python
df1=pd.DataFrame(x_train, columns=['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10','P11','P12','P13','P14','P15','P16','P17','P18','P19','P20'])
print(df1)
```

```python
#for y_train, I took 0 to 199 index rows of label column
y_train=df.iloc[0:200,1]
y_train=np.array(y_train) 
print(y_train)
print('number of training labels:', len(y_train))
```

```python
#creating dataframe
df1['Y']=y_train
df1
```

```python
alpha = 0.01
num_iters = 1000

# Initialize coefficients to zero
coeffs = np.zeros(x_train.shape[1])

# Define sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Perform gradient descent
for i in range(num_iters):
    # Calculate predicted values
    y_pred = sigmoid(np.dot(x_train, coeffs))
    
    # Calculate errors
    errors = y_pred - y_train
    
    # Calculate gradients
    gradients = np.dot(x_train.T, errors) / len(y_train)
    
    # Update coefficients
    coeffs -= alpha * gradients

# Print the coefficients
print("Coefficients: ", coeffs)
```
